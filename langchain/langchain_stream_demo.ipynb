```python
# https://docs.langchain.com/oss/python/integrations/chat/ollama
from langchain.agents import create_agent
from langchain_ollama import ChatOllama

model = ChatOllama(model="qwen3:4b", reasoning=False)
tool = {"type": "web_search"}
model_with_tools = model.bind_tools([tool])

agent = create_agent(
    model=model,
    tools=[],
    system_prompt="You are a helpful assistant",
)

inputs = {"messages": [("user", "通过web_search工具，帮我总结一下今年的重要事件。")]}

# Use stream_mode="messages" for token-level streaming
for msg, metadata in agent.stream(inputs, stream_mode="messages"):
    if msg.content:
        print(msg.content, end="", flush=True)
```
